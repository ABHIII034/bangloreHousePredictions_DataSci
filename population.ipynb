```python
import matplotlib.pyplot as plt

# Example data (Year vs Population in billions)
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
population = [2.5, 3.0, 3.7, 4.4, 5.3, 6.1, 6.9, 7.8]  # in billions

plt.figure(figsize=(10, 6))
plt.plot(years, population, marker='o', color='green', linestyle='-')

plt.title("World Population Over Time", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Population (Billions)", fontsize=14)
plt.grid(True)

plt.show()
```

```python
import numpy as np
mean=np.mean(population)
```

```python
varience=np.var(population)
```

```python
std_devaiation=np.std(population)
```

```python
print(mean)
```

```python
print(varience)
```

```python
print(std_devaiation)
```

```python
# Bar chart
stats = [mean, std_devaiation, varience]
labels = ["Mean", "Standard Deviation", "Variance"]

plt.figure(figsize=(8, 5))
plt.bar(labels, stats, color=['blue', 'orange', 'green'])

plt.title("Mean, Standard Deviation, and Variance of Population Data", fontsize=14)
plt.ylabel("Population (Millions)", fontsize=12)
plt.show()
```
